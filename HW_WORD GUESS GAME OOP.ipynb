```python
import random
from IPython.display import clear_output as co


class Word:
    words = ['program', 'coding', 'python', 'computer', 'keyboard', 'webcam']
    
    def __init__(self):
        self.known = random.choice(self.words)
        self.to_guess = ['_' for i in list(self.known)]
        
    def __str__(self):
        return self.known

class Game:
    letter_list = []
    guesses = 7
    
    @classmethod
    def game_over(self, w):
        print(f"You did not guess the word. The correct word was: {w}")
        
    @classmethod
    def game_won(self):
        print("Congratulations! You've guessed the word!")
    
    @classmethod
    def run(self):
        word = Word()
        
        while True:
            confirm = input("Press 'ENTER' to continue. Type 'Quit' to exit the program. ").lower()
            if confirm == 'quit':
                break
            if '_' not in word.to_guess:
                Game.game_won()
                break
            if self.guesses == 0:
                Game.game_over(word.known)
                break
            co()
            
            print(' '.join(word.to_guess))
            
            print('=' * 50)
            print(f'Guesses remaining: {self.guesses}')
            guess = input('Guess a letter: ')
            
            if guess.isalpha():
                if guess in word.known:
                    for idx, letter in enumerate(word.known):
                        if letter == guess:
                            word.to_guess[idx] = guess
                else:
                    self.guesses -= 1
            else:
                co()
                print('That is invalid input. Try again.')
                continue
        
Game.run()
```
